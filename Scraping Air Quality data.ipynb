```python
import selenium
from selenium import webdriver
```

```python
import pandas as pd
```

```python
driver = webdriver.Chrome(executable_path='/Users/kasturipananjady/Desktop/chromedriver2')
driver.get('http://aqicn.org/map/india/')
```

```python
list_test = driver.find_elements_by_tag_name("td")
```

```python
list_test
```

```python
list_name = []
for item in list_test:
    list_name += [item.text]  
```

```python
list_name
```

```python
list_day = []
for item in list_test:
    list_day += [item.get_attribute('title')]  
```

```python
list_day
```

```python
list_dicts = []
for item in list_test:
    dict_aqi = {}
    day = item.get_attribute('title')
#     print(item.text)
#     print(day)
    if item.text != '':
        dict_aqi['day'] = day
        dict_aqi['AQI'] = item.text
        list_dicts += [dict_aqi]
   
```

```python
list_dicts
```

```python
df1 = pd.DataFrame(list_dicts, 
                      columns =['day', 'AQI'])
```

```python
df1.to_csv('AQI.csv', index=False)
```
